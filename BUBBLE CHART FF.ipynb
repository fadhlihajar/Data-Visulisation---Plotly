```python
import plotly.plotly as py
from plotly.grid_objs import Grid, Column
from plotly.tools import FigureFactory as FF 
import pandas as pd
import time

url = 'https://raw.githubusercontent.com/fadhlihajar/DV/master/FF.csv'
dataset = pd.read_csv(url)
```

```python
years_from_col = set(dataset['year'])
years_ints = sorted(list(years_from_col))
years = [str(year) for year in years_ints]

Regions = []
for Region in dataset['Region']:
    if Region not in Regions: 
        Regions.append(Region)
        columns = []
```

```python
 for year in years:
    for Region in Regions:
        dataset_by_year = dataset[dataset['year'] == int(year)]
        dataset_by_year_and_cont = dataset_by_year[dataset_by_year['Region'] == Region]
        for col_name in dataset_by_year_and_cont:
           
            column_name = '{year}_{Region}_{header}_gapminder_grid'.format(
                year=year, Region=Region, header=col_name
            )
            a_column = Column(list(dataset_by_year_and_cont[col_name]), column_name)
            columns.append(a_column)
```

```python
grid = Grid(columns)
url = py.grid_ops.upload(grid, 'gapminder_grid'+str(time.time()), auto_open=False)
url
table = FF.create_table(dataset.head(10))
py.iplot(table, filename='FF')
```

```python
for year in years:
    frame = {'data': [], 'name': str(year)}
    for Region in Regions:
        data_dict = {
            'xsrc': grid.get_column_reference(col_name_template.format(
                year=year, Region=Region, header='lifeExp'
            )),
            'ysrc': grid.get_column_reference(col_name_template.format(
                year=year, Region=Region, header='gdpPercap'
            )),
            'mode': 'markers',
            'textsrc': grid.get_column_reference(col_name_template.format(
                year=year, Region=Region, header='country'
                )),
            'marker': {
                'sizemode': 'area',
                'sizeref': 200000,
                'sizesrc': grid.get_column_reference(col_name_template.format(
                    year=year, Region=Region, header='pop'
                )),
                'color': custom_colors[Region]
            },
            'name': Region
        }
        frame['data'].append(data_dict)

    figure['frames'].append(frame)
    slider_step = {'args': [
        [year],
        {'frame': {'duration': 300, 'redraw': False},
         'mode': 'immediate',
       'transition': {'duration': 300}}
     ],
     'label': year,
     'method': 'animate'}
    sliders_dict['steps'].append(slider_step)

figure['layout']['sliders'] = [sliders_dict]
```
